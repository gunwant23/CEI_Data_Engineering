#Missing Values Handling

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)


print("columns with most missing values")
print(pd.DataFrame({'Missing Count' : missing,
                    'Missing %'     : missing_pct
}).query('`Missing Count` > 0').sort_values('Missing %', ascending=False))
```

Fixing

```python
#Videos
df.drop(columns=['videos'],inplace=True)
print('Dropped the videos column as 78.1 % values were null, which is not useful')
```

```python
#What_customers_said
df.drop(columns=['what_customers_said'], inplace=True)
print(f'57.3% missing + unstructured JSON string = not usable without parsing.')
```

Custom Filtering(user-input based)

```python
# Get unique categories from the 'category'
unique_categories = df['category'].unique()
category_map = {i + 1: category for i, category in enumerate(unique_categories)}
print("Available categories:")
# Display the categories to the user
for num, category in category_map.items():
    print(f"{num} - {category}")
user_input = input("\nEnter category numbers separated by spaces (e.g., 1 3): ")
selected_numbers = [int(n.strip()) for n in user_input.split() if n.strip().isdigit()]
selected_categories = [category_map[num] for num in selected_numbers if num in category_map]

if selected_categories:
        filtered_products = df[df['category'].isin(selected_categories)]
        print(f"\nProducts in selected categories ({', '.join(selected_categories)}): {len(filtered_products):,} products")
        display(filtered_products[['title', 'category', 'initial_price', 'discount']].head())
else:
        print("No valid categories selected or entered.")
```

```python
min_price = float(input("Enter minimum price: "))
max_price = float(input("Enter maximum price: "))

price_range_products = df[(df['initial_price'] >= min_price) & (df['initial_price'] <= max_price)]
print(f'Products in price range ₹{min_price:.2f} - ₹{max_price:.2f}: {len(price_range_products):,} products')
price_range_products[['title', 'category', 'initial_price', 'discount']].head(5)
```

#Remove Duplicates

```python
before = len(df)
dup_count = df.duplicated().sum()
print(f'Duplicate rows (exact match) : {dup_count}')

# Check for duplicate product IDs (same product listed twice)
dup_ids = df.duplicated(subset=['product_id']).sum()
print(f'Duplicate product_id entries : {dup_ids}')

# Remove all duplicates (keep first occurrence)
df.drop_duplicates(inplace=True)
df.drop_duplicates(subset=['product_id'], keep='first', inplace=True)

after = len(df)
print(f'\nRows before : {before:,}')
print(f'Rows after  : {after:,}')
print(f'Removed     : {before - after} duplicate rows')
```

No duplicates :)

#Derived columns

```python
mean_price = df['initial_price'].mean()
df['price_tier'] = pd.cut(
    df['initial_price'],
    bins   = [0, mean_price * 0.5, mean_price, mean_price * 1.5, float('inf')],
    labels = ['Budget', 'Mid-Range', 'Premium', 'Luxury']
)
df[['title', 'category', 'initial_price', 'price_tier']].head()
```

```python
inconsistent = df[
    (df['discount'] > 0) &
    (df['final_price'] >= df['initial_price'])
]
print(f'Inconsistent rows: {len(inconsistent)}')
```

_ This Shows that even after applying discount the discount was not applied to final price

```python
# Identify inconsistent rows
inconsistent_indices = df[(df['discount'] > 0) & (df['final_price'] >= df['initial_price'])].index

# Recalculate final_price for these inconsistent rows, rounding before converting to int
df.loc[inconsistent_indices, 'final_price'] = (df.loc[inconsistent_indices, 'initial_price'] * (1 - df.loc[inconsistent_indices, 'discount'] / 100)).round().astype(int)

# Verify the fix
inconsistent_after_fix = df[
    (df['discount'] > 0) &
    (df['final_price'] >= df['initial_price'])
]
print(f'Inconsistent rows after fix: {len(inconsistent_after_fix)}')

# Display a sample of the fixed rows to show the change
display(df.loc[inconsistent_indices, ['title', 'initial_price', 'discount', 'final_price']].head())
```

```python
df['final_price_numeric'] = df['final_price'].astype(str).str.replace('₹', '').str.replace(',', '').str.replace('"', '').astype(float)
df['savings_amount'] = df['initial_price'] - df['final_price_numeric']
df['savings_amount'] = df['savings_amount'].clip(lower=0)
print(f'Total savings: ₹{df["savings_amount"].sum():,.2f}')
df[['title', 'category', 'initial_price', 'final_price_numeric', 'savings_amount']].head()
```
